```python
class CountSet:
    def __init__(self):
        self.entries = {}
        self.total_entry_count = 0
        
    def insert(self, entry):
        if entry not in self.entries:
            self.entries[entry] = 0
        self.entries[entry] += 1
        self.total_entry_count += 1
    
    def printTopN(self, n):
        sortedItems = sorted(self.entries.iteritems(), reverse=True, key=lambda entry: entry[1])    
        printList = ["".join([str(k), " : ", str(v)]) for k,v in sortedItems]
        print "\n".join(printList[:n])
    
    def __getitem__(self, key):
        return self.entries[key]
    
    def __repr__(self):
        sortedItems = sorted(self.entries.iteritems(), reverse=True, key=lambda entry: entry[1])    
        printList = ["".join([str(k), " : ", str(v)]) for k,v in sortedItems]
        return "\n".join(printList)
    
    def __len__(self):
        return self.total_entry_count

categories = set()
for cats in data_df['categories']:
    for cat in cats:
        categories.add(cat)


crossLinks = {}
for cat in categories:
    # each category keeps track of how many times it is listed with other categories
    crossLinks[cat] = CountSet() 
    
for cats in data_df['categories']:
    for cat in cats:
        for otherCat in cats:
            if cat != otherCat:
                crossLinks[cat].insert(otherCat)

stahp = 0
sortedCategories = sorted(crossLinks.iteritems(), reverse=True, key=lambda entry: len(entry[1])) 
# print the cateogories that occur together (which ones are distinguishing?)
for category, similar_set in sortedCategories:
    print category, "has {} occurrences ~~~\n".format(len(similar_set)), similar_set, "\n\n"
    stahp += 1
    if stahp > 0:
        break
```

```python
from itertools import islice
import re

class CountSet:
    def __init__(self):
        self.entries = {}
        self.total_entry_count = 0
        
    def insert(self, entry):
        if entry not in self.entries:
            self.entries[entry] = 0
        self.entries[entry] += 1
        self.total_entry_count += 1
    
    def printTopN(self, n):
        sortedItems = sorted(self.entries.iteritems(), reverse=True, key=lambda entry: entry[1])    
        printList = ["".join([str(k), " : ", str(v)]) for k,v in sortedItems]
        print "\n".join(printList[:n])
    
    def __getitem__(self, key):
        return self.entries[key]
    
    def __repr__(self):
        sortedItems = sorted(self.entries.iteritems(), reverse=True, key=lambda entry: entry[1])    
        printList = ["".join([str(k), " : ", str(v)]) for k,v in sortedItems]
        return "\n".join(printList)
    
    def __len__(self):
        return self.total_entry_count

def normalize(text):
    text = text.lower()
    text = re.sub("[^\w -]", "", text) # remove all but alphanumerics, dashes and spaces
    text = re.sub(" +", " ", text) # get rid of spaces that might arise from multiple special characters padded by a space getting deleted
    return text.strip()

stopWords = set(normalize(word) for word in open('stopWords2.txt','r').readlines())
print stopWords

f = open('yelp_academic_dataset_review.json', 'rb')
data = islice(f, 50000)

# remove the trailing "\n" from each line
data = map(lambda x: x.rstrip(), data)

# each element of 'data' is an individual JSON object.
# i want to convert it into an *array* of JSON objects
# which, in and of itself, is one large JSON object
# basically... add square brackets to the beginning
# and end, and have all the individual business JSON objects
# separated by a comma
data_json_str = b'[' + b','.join(data) + b']'

# now, load it into pandas
data_df = pd.read_json(data_json_str)
#data_df.head(10)

goodWords = CountSet()
badWords = CountSet()
for review in data_df.itertuples():
    stars = review[4]
    reviewText = review[5]
    if stars > 3:
        for token in reviewText.split():
            if token not in stopWords:
                goodWords.insert(normalize(token))
    elif stars < 3:
        for token in reviewText.split():
            if token not in stopWords:
                badWords.insert(normalize(token))  

goodWords.printTopN(20)
print "\n----------\n"
badWords.printTopN(20)
```
