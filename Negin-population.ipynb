```python
import pandas as pd

#yelp_business = pd.read_json('yelp_academic_dataset_business.json')


# read the entire file into a python array
with open('yelp_academic_dataset_business.json', 'rb') as f:
    data = f.readlines()

# remove the trailing "\n" from each line
data = map(lambda x: x.rstrip(), data)

# each element of 'data' is an individual JSON object.
# i want to convert it into an *array* of JSON objects
# which, in and of itself, is one large JSON object
# basically... add square brackets to the beginning
# and end, and have all the individual business JSON objects
# separated by a comma
data_json_str = b'[' + b','.join(data) + b']'

# now, load it into pandas
data_df = pd.read_json(data_json_str)
```

```python
rest_data_df = data_df[['Restaurants' in x for x in data_df['categories']]]
az_rest_data_df = rest_data_df[['AZ' in x for x in rest_data_df['state']]]

```

```python
import numpy as np

zip_code = [];


for index,rest in az_rest_data_df.iterrows():
    address = rest['full_address'].split('AZ ')
    if len(address) == 2:
        zip_code.append(int(address[1]))
    else:
        zip_code.append(np.nan)

az_rest_data_df['zip_code'] = zip_code


```

```python
colnames = ['zip','total population']
census_data = pd.read_csv('ACS_14_5YR_DP05.csv',
                   names=colnames, sep=',', quoting=1)
census_data = census_data[1::4]

zips = [];

for row in census_data['zip']:
    address = row.split('A5')
    zips.append(int(address[1]))

census_data['zip_code'] = zips

census_data = census_data.reset_index(drop=True)
census_data.head()

```

```python
#census_data.to_csv('census.csv', sep=',')
```

```python
frame = census_data.iloc[:,(1,2)]
az_rest_data = pd.merge(az_rest_data_df,frame,how ='left', on = 'zip_code')

az_rest_data.to_csv('arizona_restaurants_vs_population.csv', sep = ',')
az_rest_data.head()
```
