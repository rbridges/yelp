```python
import pandas as pd

#yelp_business = pd.read_json('yelp_academic_dataset_business.json')


# read the entire file into a python array
with open('yelp_academic_dataset_business.json', 'rb') as f:
    data = f.readlines()

# remove the trailing "\n" from each line
data = map(lambda x: x.rstrip(), data)

# each element of 'data' is an individual JSON object.
# i want to convert it into an *array* of JSON objects
# which, in and of itself, is one large JSON object
# basically... add square brackets to the beginning
# and end, and have all the individual business JSON objects
# separated by a comma
data_json_str = b'[' + b','.join(data) + b']'

# now, load it into pandas
data_df = pd.read_json(data_json_str)
data_df.head(10)
```

```python
import json

from jinja2 import Template
from IPython.core.display import HTML, Javascript

def init_maps():
    HTML("""
        <script async defer
            src="https://maps.googleapis.com/maps/api/js?key=AIzaSyBunQHpEUVFINnt7YvGWU4MtZLctVoq2Xo&libraries=visualization&callback=initMap">
        </script>
    """)

def heatmap_points(points, points2):
    pos = lambda x: {"lat": x[0], "lng": x[1], "w":x[2]}
    center = points.mean(axis=0)
    
    html = Template("""
        <div id="floating-panel">
            <button onclick="toggleHeatmap()">Toggle Heatmap</button>
              <button onclick="changeGradient()">Change gradient</button>
              <button onclick="changeRadius()">Change radius</button>
              <button onclick="changeOpacity()">Change opacity</button>
        </div>
        <div id="heatmap" style="height: 500px" class="map-canvas"></div>
        <script>
            var map, heatmap, heatmap2;
        
            function initMap(){
                map = new google.maps.Map(document.getElementById('heatmap'), {
                    zoom: 8,
                    center: {{center}},
                });

                heatmap = new google.maps.visualization.HeatmapLayer({
                    data: {{data}}.map(function(x) { return new google.maps.LatLng(x[0], x[1]); }),
                    map: map
                });
                heatmap.set('radius', 8);
                heatmap.set('dissipating', true);
                heatmap.set('maxIntensity', 10);
                
                heatmap2 = new google.maps.visualization.HeatmapLayer({
                    data: {{data2}}.map(function(x) { return new google.maps.LatLng(x[0], x[1]); }),
                    map: map
                });
                heatmap2.set('radius', 8);
                heatmap2.set('dissipating', true);
                heatmap2.set('maxIntensity', 10);
                var gradient = [
                  'rgba(0, 255, 255, 0)',
                  'rgba(0, 255, 255, 1)',
                  'rgba(0, 191, 255, 1)',
                  'rgba(0, 127, 255, 1)',
                  'rgba(0, 63, 255, 1)',
                  'rgba(0, 0, 255, 1)',
                  'rgba(0, 0, 223, 1)',
                  'rgba(0, 0, 191, 1)',
                  'rgba(0, 0, 159, 1)',
                  'rgba(0, 0, 127, 1)',
                  'rgba(63, 0, 91, 1)',
                  'rgba(127, 0, 63, 1)',
                  'rgba(191, 0, 31, 1)',
                  'rgba(255, 0, 0, 1)'
                ]
                heatmap2.set('gradient', gradient);
                heatmap2.set('opacity', 1.0);
            }
            
            function toggleHeatmap() {
                heatmap2.setMap(heatmap2.getMap() ? null : map);
            }
            
            function changeRadius() {
                heatmap2.set('radius', heatmap2.get('radius') ? null : 8);
            }

            function changeOpacity() {
                heatmap.set('opacity', heatmap.get('opacity') ? null : 0.3);
            }
            
        </script>
        <script async defer
            src="https://maps.googleapis.com/maps/api/js?key=AIzaSyBunQHpEUVFINnt7YvGWU4MtZLctVoq2Xo&libraries=visualization&callback=initMap">
        </script>
    """).render(
        center=json.dumps(pos(center)),
        data=json.dumps(points.tolist()),
        data2=json.dumps(points2.tolist())
    )
    return html
```

```python
import numpy as np

heat_data = []
heat_data2 = []
hdx = []
hdy = []

for index, row in rest_data_df.iterrows():
    if row['state'] == "AZ" and 'Restaurants' in row['categories'] and row['review_count'] > 33:
        heat_data.append([row['latitude'], row['longitude'], row['stars']])
        
for index, row in rest_data_df.iterrows():
    if row['state'] == "AZ" and 'Restaurants' in row['categories'] and row['stars'] >= 4.0 and row['review_count'] > 33:
        heat_data2.append([row['latitude'], row['longitude'], row['stars']])
        hdx.append(row['longitude'])
        hdy.append(row['latitude'])
```

```python
import numpy as np

HTML(heatmap_points(np.array(heat_data), np.array(heat_data2)))
```

```python
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

import numpy as np
import numpy.random
import matplotlib.pyplot as plt

heatmap, xedges, yedges = np.histogram2d(hdx, hdy, bins=75)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

heatmap = np.rot90(heatmap)
heatmap = np.flipud(heatmap)

Hmasked = np.ma.masked_where(heatmap==0,heatmap)

plt.clf()
plt.pcolormesh(xedges, yedges, Hmasked)
plt.imshow(heatmap, extent=extent)
plt.show()

print(min(hdx))
print(min(hdy))
print(heatmap.shape)
print(heatmap)
```

```python
num_pcorrect = 0;
num_pwrong = 0;

num_dcorrect = 0;
num_dwrong = 0;

for index, row in rest_data_df.iterrows():
    if row['state'] == "AZ" and 'Restaurants' in row['categories'] and row['review_count'] > 33:
        if heatmap[int((row['longitude'] - min(hdx)) * (75 / (max(hdx) - min(hdx))) % 75), int((row['latitude'] - min(hdy)) * (75 / (max(hdy) - min(hdy))) % 75)] >= 0.1:
            if row['stars'] >= 4.0:
                num_pcorrect += 1;
            else:
                num_pwrong += 1;
        else:
            if row['stars'] < 4.0:
                num_dcorrect += 1;
            else:
                num_dwrong += 1;
                
print(num_pcorrect)
print(num_pwrong)
print(num_dcorrect)
print(num_dwrong)
print(np.amax(heatmap))
```
